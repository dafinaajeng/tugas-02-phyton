```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.plot(curah_hujan['Temperatur_ratarata(oC)'], label='Temperatur_ratarata(oC)')
plt.plot(curah_hujan['Temperatur_max(oC)'], label='Temperatur_max(oC)')
plt.plot(curah_hujan['Temperatur_min(oC)'], label='Temperatur_min(oC)')

plt.xlabel('bulan(januari=0)')
plt.ylabel('suhu')
plt.title('grafik suhu bandung')
plt.legend()
plt.show()
```
